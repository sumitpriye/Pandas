```python
# zip always returns a tuple. It tries to make pairs
```

```python
for i in zip([1,2,3],[4,5,6]):
    print(i)
```

```python
for i in zip([1,2,3],[4,5,6],[7,8,9]):
    print(i)
```

```python
for i in zip(*values):    # * here is used to make sure zip function treats the values as a nested list .otherwise it will treat it as a single list
        print(i)
```

```python
for i in zip(values):
        print(i)
```

```python
z= zip(header,zip(*values))
```

```python
z
```

```python
for i in z:
    print(i)
```

```python
type(z)
```

```python
data_dict={h:v for h,v in zip(header,zip(*values))}  # using dictionary comprehension
```

```python
data_dict
```

```python
# Normal way without Dictionary Comprehension

d={}
for i in zip(header,zip(*values)):
    d[i[0]]=i[1]
    
d
```
