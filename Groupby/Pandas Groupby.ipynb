```python
# split-apply-combine

def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
movies.groupby('Genre').apply(foo).sort_values(ascending=False)
```

```python
# Find rank of each movie in the group according to the imdb score
```

```python
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
movies.groupby('Genre', group_keys=False).apply(rank_movie)
```

```python
# Normalize the ratings
```

```python
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

movies.groupby('Genre').apply(normal)
```

```python
# Exercise

ipl = pd.read_csv('deliveries.csv')
ipl.head()
```

```python
# Find the top ten batsman in terms of runs scored


ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
# Find the batsman with maximum no of sixes

sixes= ipl[ipl['batsman_runs']==6]
sixes.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# Find the batsman with most number of 4's and 6's in last 5 overs
```

```python
death_overs = ipl[ipl['over']>=16]
```

```python
fours_sixes =  death_overs[(death_overs['batsman_runs']==4) | (death_overs['batsman_runs']==6)]
fours_sixes.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
# Find Virat Kohli's record against all teams
```

```python
kohli = ipl[ipl['batsman']=='V Kohli']
kohli.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

```python
# Create a function that can return the highest score of any batsman

def highest_runs(batsman):
    player_runs_per_match = ipl.groupby(['batsman', 'match_id'])['batsman_runs'].sum().reset_index().sort_values(by='batsman_runs', ascending=False)
    return player_runs_per_match[player_runs_per_match['batsman']==batsman].head(1)['batsman_runs'].values[0]
```

```python
player_runs_per_match = ipl.groupby(['batsman', 'match_id'])['batsman_runs'].sum().reset_index().sort_values(by='batsman_runs', ascending=False)
player_runs_per_match
```
